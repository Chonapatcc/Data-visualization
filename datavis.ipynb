```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

\
<span style="color:#029DA3">**Line Plots**</span>

Line plot is basically used to plot the relationship between two numerical sets of values. Usually, an increasing or decreasing trend between two dependent variables. Let us plot a line plot that displays the square root of 20 equidistant numbers between 0 and 20.

\
First, we generate 11 equidistant numbers using np.linspace() function.

```python
x = np.linspace(0, 100, 11)
x
```

\
Next, we iterate through each value in the x list and take the square root of each value.

```python
y = [math.sqrt(i) for i in x]
y
```

\
Plot a line plot using plot() method.  
It is important to mention that plt is an alias for pyplot.

```python
plt.plot(x, y)
```

\
There is also another way to do line plot. You first have to call the figure() method via the plt module, which draws an empty figure. Next, you can call the axes() method, which returns an axes object. You can then call the plot() method from the axes object to create a plot.

```python
x = np.linspace(0, 100, 11)
y = [math.sqrt(i) for i in x]

fig = plt.figure()
ax = plt.axes()
ax.plot(x, y)
```

You can also increase the default plot size of a Matplotlib plot. To do so, you can use the rcParams list of the pyplot module and then set two values for the figure.figsize attribute.

```python
# before changing plot size, let's print the defalut size by
print(plt.rcParams.get('figure.figsize'))
plt.rcParams['figure.figsize'] = [5, 5]
plt.plot(x, y)
```

```python
plt.rcParams['figure.figsize'] = [6, 4]
```

\
To add labels on x and y axes, you need to pass the string values respectively to the xlabel and ylabel methods of the pyplot module. Similarly, to set the title, you need to pass a string value to the title method.

```python
x = np.linspace(0, 100, 11)
y = [math.sqrt(i) for i in x]

plt.xlabel('Values')
plt.ylabel('Sqrt')
plt.title('Square Roots')

plt.plot(x, y)
```

\
To add a legend, you need to make two changes. First, you have to pass a string value for the label parameter of the plot() function. Next, you have to pass the value for the loc parameter of the legend method of the pyplot module. In the loc, you have to pass the location of your legend.

```python
x = np.linspace(0, 100, 11)
y = [math.sqrt(i) for i in x]

plt.xlabel('Values')
plt.ylabel('Sqrt')
plt.title('Square Roots')

plt.plot(
    x,
    y,
    color='r',  # r = red
    label='sqrt(value)')

plt.legend(loc='center right')
# loc can be 'best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 'center right', 'lower center', 'upper center', 'center'
```

\
You can also plot multiple line plots inside one graph. All you have to do is call the plot() method twice with different values for x and y axes. 

```python
x = np.linspace(0, 100, 11)
y = [math.sqrt(i) for i in x]
y2 = x**(1 / math.e)

plt.xlabel('Values')
plt.ylabel('Sqrt')
plt.title('Square/e Roots')

#1st line
plt.plot(
    x,
    y,
    color='r',  # r = red
    label='sqrt(value)')

#2nd line
plt.plot(
    x,
    y2,
    color='b',  # r = red
    label='value^(1/e)')

plt.legend(loc='center right')
```

<span style="color:#029DA3">**Scatter Plots**</span>

A Scatter plot is used essentially to plot the relationship between two numeric columns in the form of scattered points. Normally, a scattered plot is used when for each value in the x-axis, there exist multiple values in the y-axis. To plot a scatter plot, the scatter() function of the pyplot module is used. You have to pass the values for the x-axis and y-axis. In addition, you have to pass a shorthand notation of color value to the c/color parameter. 

```python
plt.xlabel('age')
plt.ylabel('serum_creatinine')
plt.title('age & serum_creatinine'.title())

plt.scatter(
    df_heart.age,
    df_heart.serum_creatinine,
    color='r',  # r = red
    edgecolor='w', # w = white
    s=100,
    alpha=0.5)
```

For more information, see https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html

<span style="color:#029DA3">**Bar Plots**</span>

Bar plot is used to plot the relationship between unique values in a categorical column grouped by an aggregate function such as sum, mean, median, etc. To plot a bar plot, you need to call the bar() method. The categorical values are passed as the x-axis and corresponding aggregated numerical values are passed on the y-axis. The following script plots a bar plot between DEATH_EVENT and ages of the patients from the heart failure dataset.

```python
plt.xlabel('Decease')
plt.ylabel('Age')
plt.title('age vs survival'.title())

x_vals = df_heart.DEATH_EVENT.unique()
y_vals = df_heart.groupby('DEATH_EVENT').age.mean()[x_vals]

plt.bar(x=x_vals,
        height=y_vals,
        width=0.5,
        align='center',
        color=["r", "b"],
        edgecolor="b",
        alpha=0.7)
```

\
To add legend, we need to add plt.legend and specify argument handles and labels.

```python
plt.xlabel('Decease')
plt.ylabel('Age')
plt.title('age vs survival'.title())

x_vals = df_heart.DEATH_EVENT.unique()
y_vals = df_heart.groupby('DEATH_EVENT').age.mean()[x_vals]

plt.bar(x=x_vals,
        height=y_vals,
        width=0.5,
        align='center',
        color=["r", "b"],
        edgecolor="b",
        alpha=0.7)

colors = {'Y': 'r', 'N': 'b'}
labels = x_vals
handles = [
    plt.Rectangle(xy=(0, 0), width=1, height=1, color=colors[label])
    for label in labels
]
plt.legend(handles, labels, loc='upper center')
```

\
The following is another example bar plot between DEATH_EVENT and follow up time.

```python
plt.xlabel('Decease')
plt.ylabel('Total follow up (days)')
plt.title('Decease vs days follow up'.title())

x_vals = df_heart.DEATH_EVENT.unique()
y_vals = df_heart.groupby('DEATH_EVENT').time.mean()[
    df_heart.DEATH_EVENT.unique()]

plt.bar(x=x_vals,
        height=y_vals,
        width=0.5,
        align='center',
        color=["r", "b"],
        edgecolor="b",
        alpha=0.7)

colors = {'Y': 'r', 'N': 'b'}
labels = x_vals
handles = [
    plt.Rectangle(xy=(0, 0), width=1, height=1, color=colors[label])
    for label in labels
]
plt.legend(handles, labels, loc='upper center')
```

<span style="color:#029DA3">**Pie Charts**</span>

Pie charts, as the name suggests, display the percentage distribution of values in a categorical column in terms of an aggregated function. For instance, the following script shows the percentage distribution of categories according to the Pclass column of titanic dataset.  
\
To plot a pie chart, the pie() method of the pyplot module is used. The first parameter is the list of the numeric values that you want converted and displayed into percentages. Next, you have to pass a list of categories to the labels parameter. The explode parameter defines the magnitude of the split for each category in the pie chart. The autopct parameter defines the format in which the percentage will be displayed on the pie chart.

```python
df_titanic = pd.read_csv("https://raw.githubusercontent.com/ThammakornS/ProgStat/main/dataset/titanic.csv")
df_titanic.head()
```

```python
plt.title('Passenger Classes')

labels = df_titanic.Pclass.unique()
values = df_titanic.Pclass.value_counts() / df_titanic.shape[0] * 100
explode = [0.05] * df_titanic.Pclass.unique().shape[0]

plt.pie(x=values,
        explode=explode,
        labels=labels,
        autopct='%1.1f%%',
        shadow=True)
```

For more information, see https://matplotlib.org/stable/gallery/pie_and_polar_charts/pie_features.html

<span style="color:#029DA3">**Subplots**</span>

The subplots() is used to enclose multiple plots into a single set. We can specify the subplot dimension using nrows and ncols arguments.  
To put seaborn's plots into the subplot, specify parameter 'ax'.  
To put general matplotlib's plots into the subplot, assign plot object to 'axes' variable. 

```python
fig, axes = plt.subplots(1, 2, figsize=(10,5))  #1 row, 2 columns

fig.suptitle('Titanic Passenger Class')

# row 0 column 0 plot
sns.boxplot(ax=axes[0], # To put seaborn's plots into the subplot
            data=df_titanic,
            x='Pclass',
            y='Age',
            hue='Survived',
            width=0.5,
            whis=1.5,
            linewidth=0.5,
            fliersize=1)

# row 0 column 1 plot
labels = df_titanic.Pclass.unique()
values = df_titanic.Pclass.value_counts() / df_titanic.shape[0] * 100
explode = [0.05] * df_titanic.Pclass.unique().shape[0]

#To put general matplotlib's plots into the subplot
axes[1] = plt.pie(x=values, #To put general matplotlib's plots into the subplot
                  explode=explode,
                  labels=labels,
                  autopct='%1.1f%%',
                  shadow=True)
```

!NOTE: If you create a subplot with 2d dimension like   
fig, axes = plt.subplots(2, 2)  
When use axes, you also need to specify in 2d style.
For example,  
ax=axes[0, 0]  
axes[0, 1] = plt.pie(...)

<span style="color:#029DA3">**Exporting Plots**</span>

In general, we will export our plot as a file for further uses such as publication or preparing presentation.  
Matplotlib already provides the savefig() function for this task.

- Save as PNG (Portable Network Graphics) file

```python
sns.set_style("darkgrid")

sns.catplot(data=df_heart,
             kind='box',
             col='sex',
             x='DEATH_EVENT',
             y='time',
             hue='diabetes',
             width=0.5,
             whis=1.5,
             linewidth=0.5,
             fliersize=1)

plt.savefig('plot.png',
           dpi=70, # for setting resolution
           transparent=True) # make transparent png
```

- Save as pdf / eps / svg file

```python
sns.set_style("darkgrid")

sns.catplot(data=df_heart,
             kind='box',
             col='sex',
             x='DEATH_EVENT',
             y='time',
             hue='diabetes',
             width=0.5,
             whis=1.5,
             linewidth=0.5,
             fliersize=1)

plt.savefig('plot.pdf') #for pdf
#plt.savefig('plot.eps') #for eps
#plt.savefig('plot.svg') #for svg
```

- Save as tiff file

```python
sns.set_style("darkgrid")

sns.catplot(data=df_heart,
             kind='box',
             col='sex',
             x='DEATH_EVENT',
             y='time',
             hue='diabetes',
             width=0.5,
             whis=1.5,
             linewidth=0.5,
             fliersize=1)

plt.savefig('plot.tiff',
           dpi=70) 
```
